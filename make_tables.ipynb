```python
import pandas as pd
```

```python
df = pd.read_csv("classification.csv",sep="\t")

df
```

```python
df_tmp_forecast_forecast = df.drop(columns=["loader","\\lambda","\\tau_{inst}"])

# rename if the name exists in the dictionary
name_map= { i:"$"+i+"$" for i in ["\\tau_{temp}","\\tau_{inst}","\\lambda"]}
df_tmp_forecast_forecast.rename(columns = name_map,inplace=True)

print(df_tmp_forecast_forecast.to_latex(index=False,
                  formatters={"\\tau_{temp}": "{:,.2f}".format,"\\tau_{inst}": "{:,.2f}".format,},
                  float_format="{:,.3f}".format
))  
```
